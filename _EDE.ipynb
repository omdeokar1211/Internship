```python
df = sns.load_dataset('iris') 
df
```

```python
# Pairplot 

sns.pairplot(df, hue='species') 
plt.show() 
```

```python
# Drop Categorical column

df_numeric = df.drop(columns='species')   

df_numeric = df.drop('species', axis=1)    
correlation_matrix = df_numeric.corr()
correlation_matrix
```

```python
# Heatmap

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm') 
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Pie chart  

labels = ['A', 'B', 'C'] 
sizes = [30, 50, 20] 

plt.pie(sizes, labels=labels, autopct='%1.1f%%',startangle =90)    
plt.title("Pie Chart") 
plt.show() 
```

```python
# Strip plot

sns.stripplot(x='species', y='sepal_length', data=df, jitter=True) 
plt.title("Strip Plot") 
plt.show()
```

```python
# Joint plot

sns.jointplot(x='sepal_length', y='petal_length', data=df, kind='scatter') 
plt.show()
```

```python
# Rug plot

sns.rugplot(data) 
plt.title("Rug Plot") 
plt.show() 
```

```python
# Violin plot 

sns.violinplot(y=data) 
plt.title("Violin Plot") 
plt.show() 
```
