```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(data=ds,x='Department',palette='Set2')
plt.title('Employeee count per Department')
```

Finance department having more empolyee count followed by marketing and engineering

```python
#gender distribution
sns.countplot(data=ds,x='Gender',palette='coolwarm')
plt.title('Gender Distribution')
```

Femele employee count is more then mele count

```python
#employee location distribution
sns.countplot(data=ds,x='Location',palette='Spectral')
plt.title('Number of employeee by location')
plt.xticks(rotation=45)

```

number of employee is seatle is more follwed by new yark and San Francisco

```python
#Age Distribution

sns.histplot(ds['Age'],bins=10,kde=True)
plt.title('Age distribution')
```

employee in 40-45 years are more followed by 25-30

```python
sns.histplot(ds['Years_of_Experience'],bins=15,kde=True)
plt.title('Years_of_Experience')
```

2.5-5.0 years experience employeee are more then other years of experience employes 

```python
sns.boxplot(data=ds, y='Monthly_Salary', color='lightblue')
plt.title('Salary Seprated')
```

20000 is minimum salary and maximum salary is distributed and 750000 is average salary for distributed

```python
#average salary by department
sns.barplot (data=ds,x='Department', y='Monthly_Salary', estimator=np.mean , palette='viridis')
plt.title("average salary by department")

             
```

Finance department having more avarage Salary

```python
sns.boxplot(data=ds, x='Department', y='Performance_Score', palette='pastel')
plt.title('Salary by Gender')
```

```python
#Salary Distibution by Gender
sns.violinplot(data=ds, x='Gender', y='Monthly_Salary', palette='muted')
plt.title('Salary by Gender')
```

```python
sns.heatmap(ds[['Age','Years_of_Experience','Monthly_Salary','Performance_Score']].corr(),annot=True,cmap='coolwarm')
plt.title('Correlation matrix')
```

```python
sns.scatterplot(data=ds, x='Years_of_Experience',y='Monthly_Salary', hue='Department')
plt.title('Salery vs Exp by dept')
```
